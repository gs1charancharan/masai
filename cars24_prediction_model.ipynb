```python
import pandas as pd
```

```python
df = pd.read_csv("cars24-car-price-cleaned-new.csv")
df.head()
```

```python
df['make'] = df.groupby('make')['selling_price'].transform('mean')
df['model'] = df.groupby('model')['selling_price'].transform('mean')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
df
```

```python
from sklearn.model_selection import train_test_split
df_train,df_test = train_test_split(df,test_size=0.3,random_state=1)

df_train.shape,df_test.shape
```

```python
y_train = df_train['selling_price']
X_train = df_train.drop('selling_price',axis=1)

y_train.shape,X_train.shape

X1_train = X_train[['model']]
```

```python
y_test = df_test['selling_price']
X_test = df_test.drop('selling_price', axis=1)

y_test.shape,X_test.shape

#X1_test = X_test[['model']]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train,y_train)
```

```python
y_hat = model.predict(X_test)
y_hat
```

```python
from sklearn.dummy import DummyRegressor

baseline_regressor = DummyRegressor(strategy="mean")
baseline_regressor.fit(X_train, y_train)

baseline_predictions = baseline_regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("--- Baseline Model Metrics ---")
print(f"Baseline MAE: {mean_absolute_error(y_test, baseline_predictions):.4f}")
print(f"Baseline MSE: {mean_squared_error(y_test, baseline_predictions):.4f}")
print(f"Baseline R2: {r2_score(y_test, baseline_predictions):.4f}")

print("\n--- ML Model Metrics (Univariate Linear Regression) ---")
print(f"ML Model MAE: {mean_absolute_error(y_test, y_hat):.4f}")
print(f"ML Model MSE: {mean_squared_error(y_test, y_hat):.4f}")
print(f"ML Model R2: {r2_score(y_test, y_hat):.4f}")
```
