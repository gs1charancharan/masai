You are a machine learning engineer implementing a simple linear regression model from scratch using NumPy to predict house prices. You need to implement gradient descent to optimize the model parameters.

**Problem Statement:**
Given a dataset of house areas (in sq.ft) and their corresponding prices (in lakhs), implement gradient descent to find optimal weight (w) and bias (b) for the linear model: price = w * area + b


```python
import numpy as np
areas = np.array([500, 800, 1200, 1500, 2000], dtype=np.float64)
prices = np.array([20, 30, 45, 55, 75], dtype=np.float64)
```

**Implementation Requirements:**
1. Initialize weight w = 0.0 and bias b = 0.0
2. Set learning_rate = 0.00001 and num_iterations = 1000
3. In each iteration:
   - Calculate predictions using: predictions = w * areas + b (use NumPy vectorization)
   - Calculate Mean Squared Error (MSE) loss: loss = (1/n) * sum((predictions - prices)Â²) where n is number of samples
   - Compute gradients:
     * dw = (2/n) * sum((predictions - prices) * areas)
     * db = (2/n) * sum(predictions - prices)
   - Update parameters:
     * w = w - learning_rate * dw
     * b = b - learning_rate * db
4. After training, predict the price for a new house with area = 1800 sq.ft

```python
w=0.0
b=0.0
lr=0.00001
num_iter=1000
```

```python
def mse(pred,prices):
  n=prices.shape[0]
  return (1.0/n) * np.sum((pred-prices)**2)

n=areas.shape[0]

pred= w * areas + b
init_loss=mse(pred,prices)
print(w,b)
```

```python
for _ in range(num_iter):
  dw=(2.0/n) * np.sum((pred-prices)*areas)
  db=(2.0/n) * np.sum(pred-prices)
  w=w-lr*dw
  b=b-lr*db
```

```python
pred=w*areas+b
prediction_1800=w*1800+b
```

**Output Requirements:**
Print the following with proper formatting:
- Line 1: 'Initial Loss: X.XX' (loss at iteration 0, 2 decimal places)
- Line 2: 'Final Loss: X.XX' (loss after 1000 iterations, 2 decimal places)
- Line 3: 'Optimized Weight: X.XXXX' (final w value, 4 decimal places)
- Line 4: 'Optimized Bias: X.XXXX' (final b value, 4 decimal places)
- Line 5: 'Predicted Price for 1800 sq.ft: Rs. X.XX lakhs' (2 decimal places)

```python
print(f"Initial Loss: {init_loss:.2f}")
print(f"Final Loss: {mse(pred,prices):.2f}")
print(f"Optimized Weight: {w:.4f}")
print(f"Optimized Bias: {b:.4f}")
print(f"Predicted Price for 1800 sq.ft is {prediction_1800:.2f}")
```
